Welcome to New York City, one of the most-visited cities in the world. There are many Airbnb listings in New York City to meet the high demand for temporary lodging for travelers, which can be anywhere between a few nights to many months. In this project, we will take a closer look at the New York Airbnb market by combining data from multiple file types like `.csv`, `.tsv`, and `.xlsx`.

Recall that **CSV**, **TSV**, and **Excel** files are three common formats for storing data. 
Three files containing data on 2019 Airbnb listings are available to you:

**data/airbnb_price.csv**
This is a CSV file containing data on Airbnb listing prices and locations.
- **`listing_id`**: unique identifier of listing
- **`price`**: nightly listing price in USD
- **`nbhood_full`**: name of borough and neighborhood where listing is located

**data/airbnb_room_type.xlsx**
This is an Excel file containing data on Airbnb listing descriptions and room types.
- **`listing_id`**: unique identifier of listing
- **`description`**: listing description
- **`room_type`**: Airbnb has three types of rooms: shared rooms, private rooms, and entire homes/apartments

**data/airbnb_last_review.tsv**
This is a TSV file containing data on Airbnb host names and review dates.
- **`listing_id`**: unique identifier of listing
- **`host_name`**: name of listing host
- **`last_review`**: date when the listing was last reviewed

```python
# Import the numpy and pandas packages
import numpy as np
import pandas as pd
import datetime as dt

# Load the df_price, df_room_type and df_last_review dataframes using appropriate pandas methods
df_price = pd.read_csv("data/airbnb_price.csv")
df_room_type = pd.read_excel("data/airbnb_room_type.xlsx").to_csv("data/airbnb_room_type.csv", index = None, header = True)
df_room_type = pd.read_csv("data/airbnb_room_type.csv")
df_last_review = pd.read_csv("data/airbnb_last_review.tsv", sep = "\t")
```

```python
# Print the head of the df_price dataframe
print(df_price.head())
```

```python
# Print the head of the df_room_type dataframe
print(df_room_type.head())
```

```python
# Print the head of the df_last_review dataframe
print(df_last_review.head())
```

```python
# Convert the "last_review" column of the df_last_review dataframe
df_last_review["last_review"] = pd.to_datetime(df_last_review["last_review"])

# Extract the earliest and latest review dates from the df_last_review dataframe and store the result in the earliest_review_date and latest_review_date variables, respectively, and print the result
earliest_review_date = df_last_review["last_review"].min()
latest_review_date = df_last_review["last_review"].max()
print("Earliest review date: ", earliest_review_date.strftime("%d/%m/%Y"))
print("Latest review date: ", latest_review_date.strftime("%d/%m/%Y"))
```

```python
# Clean the "room_type" column of the df_room_type dataframe by converting all entries in the column to lowercase format, so as to have a unique possible value attained by each room type
df_room_type["room_type"] = df_room_type["room_type"].str.lower()

# Count the number of listings which are private rooms using the df_room_type dataframe and store the result in the number_of_private_rooms variable, and print the result
number_of_private_rooms = len(df_room_type[df_room_type["room_type"] == "private room"]["room_type"])
print("Number of private rooms: ", number_of_private_rooms)
```

```python
# Clean the "price" column of the df_price dataframe by converting all entries in the column to float format, so as to have numerical values for each price
df_price["price"] = df_price["price"].str.replace("dollars", "").str.replace(" ", "")
df_price["price"] = df_price["price"].astype("float")

# Calculate, to the nearest penny, the average listing price from the mean of the "price" column of the df_price dataframe and store it in the average_listing_price variable
average_listing_price = round(df_price["price"].mean(), 2)
print("Average listing price: $", average_listing_price)
```

```python
# Combine all newly created variables in a new dataframe called review_dates with one row and four columns in the following order: first_reviewed, last_reviewed, nb_private_rooms, and avg_price, and print the head of the review_dates dataframe
review_dates = pd.DataFrame(index=[0], data={
    "first_reviewed": earliest_review_date,
    "last_reviewed": latest_review_date,
    "nb_private_rooms": number_of_private_rooms,
    "avg_price": average_listing_price
})
print(review_dates.head())
```
