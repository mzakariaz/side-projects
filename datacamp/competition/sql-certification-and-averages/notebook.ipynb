## üìù Step 2: Explain your choice of metrics

Your boss was impressed with the analysis work that you did and shared it with the team leaders. They have no background of working with data and don‚Äôt understand what you mean by average, or why you picked the approach you did to calculate it. 

Can you explain to the team leaders, **in no more than 200 words**, how you calculated the average, what an average means and the different approaches to averages, and why you picked this summary?

An **average** represents a dataset's "typical value". Three types are: **mean**, **median**, and **mode**.

1. **Mean:** The sum of values divided by their count. For example, the mean of $(3,2,3,4)$ is given by: 
$$\frac{3 + 2 + 3 + 4}{4} = 3$$
The mean summarizes the dataset's overall position but is sensitive to outliers.

2. **Median:** The middle value in a dataset, after sorting it in ascending order. If there's an even count, it's the mean of the two middle values. For example:
  - The median of $(3, 1, 5, 9, 7)$ is given by:
    $$(3, 1, 5, 9, 7) \xrightarrow{\text{sort}} (1, 3, 5, 7, 9) \xrightarrow{\text{middle value}} 5$$
  - The median of $(1, 3, 1, 4, 5, 0)$ is given by:
    $$(1, 3, 1, 4, 5, 0) \xrightarrow{\text{sort}} (0, 1, 1, 3, 4, 5) \xrightarrow{\text{mean of middle values}} \frac{1 + 3}{2} = 2$$
The median summarizes the dataset's central tendency without being skewed by outliers.

3. **Mode:** The most frequent value(s). For example:
  - The mode of $(3, 2, 3, 4)$ is $3$.
  - The mode of $(1, 3, 1, 4, 5, 0)$ is $1$.

The mode highlights the most common value but gives little information on the dataset's overall position.
