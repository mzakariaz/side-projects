Given the large number of movies and series available on the platform, it is a perfect opportunity to flex your exploratory data analysis skills and dive into the entertainment industry. Our friend has also been brushing up on their Python skills and has taken a first crack at a CSV file containing Netflix data. They believe that the average duration of movies has been declining. Using your friends initial research, you'll delve into the Netflix data to see if you can determine whether movie lengths are actually getting shorter and explain some of the contributing factors, if any.

You have been supplied with the dataset `netflix_data.csv` , along with the following table detailing the column names and descriptions:

## The data
### **netflix_data.csv**
| Column | Description |
|--------|-------------|
| `show_id` | The ID of the show |
| `type` | Type of show |
| `title` | Title of the show |
| `director` | Director of the show |
| `cast` | Cast of the show |
| `country` | Country of origin |
| `date_added` | Date added to Netflix |
| `release_year` | Year of Netflix release |
| `duration` | Duration of the show in minutes |
| `description` | Description of the show |
| `genre` | Show genre |

```python
# Imports the pandas and matplotlib packages
import pandas as pd
import matplotlib.pyplot as plt

# Loads the "netflix_data.csv" CSV file into a pandas dataframe and stores the result as netflix_df
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df.head()
```

```python
# Subsets the netflix_df dataframe by removing all TV shows and stores the result as netflix_subset
netflix_subset = netflix_df[netflix_df["type"] == "Movie"]
netflix_subset.head()
```

```python
# Further subsets the netflix_subset dataframe by removing all TV shows and stores the result as netflix_subset
netflix_movies = netflix_subset[["title","country","genre","release_year","duration"]]
netflix_movies.head()
```

```python
# Filters the netflix_movies dataframe by keeping only movies which are shorter than 60 minutes in duration and stores the result as short_movies
short_movies = netflix_movies[netflix_movies["duration"] < 60]
short_movies.head()
```

```python
# Assigns four colors: red, blue, yellow and green, to each row of the short-movies dataframe using loops and creates a scatter plot of the result
colors = []
L = list(netflix_movies["genre"])
l = len(L)

for i in range(l):
    if L[i] == "Children":
        colors.append("r")
    elif L[i] == "Documentaries":
        colors.append("b")
    elif L[i] == "Stand-Up":
        colors.append("y")
    elif not(L[i] in ["Children", "Documentaties", "Stand-Up"]):
        colors.append("g")

fig = plt.figure(figsize=(12,8))

plt.scatter(x = netflix_movies["release_year"], y = netflix_movies["duration"], color = colors)
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.title("Movie Duration by Year of Release")
plt.legend()
plt.show()
```
