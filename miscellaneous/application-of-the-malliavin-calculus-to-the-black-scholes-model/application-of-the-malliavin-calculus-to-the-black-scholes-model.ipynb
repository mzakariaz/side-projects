```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

<h2><b>1) Monte-Carlo Simulation of the Delta of a European Call Option</b></h2>

<h3><b>a) Theoretical Results</b></h3>
<ul>


<li>

Payoff function: $$\Phi(S_{T}) = (S_{T} - K)^{+}$$

</li>

<li>

Definition: $$\Delta = \frac{\partial V_{0}}{\partial S_{0}}$$

</li>

<li>

Analytic expression: $$\Delta = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{d}e^{-\frac{x^2}{2}}\mathrm{d}x$$
where: $$d = \frac{\ln\left(\frac{S_{0}}{K}\right) + (r + \frac{\sigma^2}{2})T}{\sigma\sqrt{T}}$$

</li>

<li>

Malliavin expression: $$\Delta = \frac{e^{-rT}}{\sigma T S_{0}}\mathbb{E}\left[\Phi(S_{T})W_{T}\right] $$

</li>

</ul>

<h3><b>b) Python Implementation</b></h3>

```python
def MonteCarloEuropeanCallDelta(r,sigma,S0,K,T,N):
    start = time.time()
    d = (np.log((S0) / (K)) + (r + (sigma**2 / 2)) * T) / (sigma * np.sqrt(T))
    SAMPLE = []
    PATH = []
    for i in range(N):
        W = np.random.normal(0, np.sqrt(T))                                             
        S = S0*np.exp((r - (sigma**2) / 2)*T + sigma*W)                                 
        V = max(S - K,0)
        SAMPLE.append(np.exp(-r*T)*V*W / (S0*T*sigma))
        PATH.append(sum(SAMPLE) / len(SAMPLE))
    end = time.time()
    print("Risk-free interest rate: r = " + str(r))
    print("Volatility: \u03C3 = " + str(sigma))
    print("Initial asset price: " + "S0".translate(str.maketrans("0", "₀")) + " = " + str(S0))
    print("Strike price: K = " + str(K))
    print("Maturity date: T = " + str(T))
    print("Number of simulations: N = " + str(N))
    print("Computation time: " + str(round(end - start, 3)) + " seconds")
    plt.rcParams["font.family"] = "Times New Roman"
    pd.Series(PATH).plot(linewidth = 0.75, color = "b")    
    plt.axhline(norm.cdf(d), color = "black", linestyle = "--", linewidth = 0.75)                         
    plt.xlabel("Number of Simulations")
    plt.ylabel("Estimated Delta")
    plt.legend(["Sample Mean", "Exact Value (" + str(round(norm.cdf(d),4)) + ")"])
    plt.title("Monte-Carlo Simulation of the Delta of a European Call Option")
    plt.rcParams['figure.dpi'] = 300
    plt.rcParams['savefig.dpi'] = 300

MonteCarloEuropeanCallDelta(0.1,0.2,100,100,1,20000)
```

<h2><b>2) Monte-Carlo Simulation of the Vega of a European Call Option</b></h2>

<h3><b>a) Theoretical Results</b></h3>
<ul>


<li>

Payoff function: $$\Phi(S_{T}) = (S_{T} - K)^{+}$$

</li>

<li>

Definition: $$\mathcal{V} = \frac{\partial V_{0}}{\partial \sigma}$$

</li>

<li>

Analytic expression: $$\mathcal{V} = S_{0}e^{-\frac{d^2}{2}}\sqrt{\frac{T}{2\pi}}$$
where: $$d = \frac{\ln\left(\frac{S_{0}}{K}\right) + (r + \frac{\sigma^2}{2})T}{\sigma\sqrt{T}}$$

</li>

<li>

Malliavin expression: $$\mathcal{V} = e^{-rT}\mathbb{E}\left[\Phi(S_{T})\left(\frac{W_{T}^2-T}{\sigma T} - W_{T}\right)\right] $$

</li>

</ul>

<h3><b>b) Python Implementation</b></h3>

```python
def MonteCarloEuropeanCallVega(r,sigma,S0,K,T,N):
    start = time.time()
    d = (np.log((S0) / (K)) + (r + (sigma**2 / 2)) * T) / (sigma * np.sqrt(T))
    SAMPLE = []
    PATH = []  
    for i in range(N):
        W = np.random.normal(0, np.sqrt(T))                                             
        S = S0*np.exp((r - (sigma**2) / 2)*T + sigma*W)                                 
        V = max(S - K,0)
        SAMPLE.append(np.exp(-r*T) * V * (((W**2 - T) / (sigma * T)) - T))
        PATH.append(sum(SAMPLE) / len(SAMPLE))
    end = time.time()
    print("Risk-free interest rate: r = " + str(r))
    print("Volatility: \u03C3 = " + str(sigma))
    print("Initial asset price: " + "S0".translate(str.maketrans("0", "₀")) + " = " + str(S0))
    print("Strike price: K = " + str(K))
    print("Maturity date: T = " + str(T))
    print("Number of simulations: N = " + str(N))
    print("Computation time: " + str(round(end - start, 3)) + " seconds")
    plt.rcParams["font.family"] = "Times New Roman"
    pd.Series(PATH).plot(linewidth = 0.75, color = "b")    
    plt.axhline(S0*np.exp(-(d**2) / 2)*np.sqrt((T) / (2*np.pi)), color = "black", linestyle = "--", linewidth = 0.75)                
    plt.xlabel("Number of Simulations")
    plt.ylabel("Estimated Vega")
    plt.legend(["Sample Mean", "Exact Value (" + str(round(S0*np.exp(-(d**2) / 2)*np.sqrt((T) / (2*np.pi)),4)) + ")"]) 
    plt.title("Monte-Carlo Simulation of the Vega of a European Call Option") 
    plt.rcParams['figure.dpi'] = 300
    plt.rcParams['savefig.dpi'] = 300

MonteCarloEuropeanCallVega(0.1,0.2,100,100,1,20000)
```

<h2><b>3) Monte-Carlo Simulation of the Gamma of a European Call Option</b></h2>

<h3><b>a) Theoretical Results</b></h3>
<ul>


<li>

Payoff function: $$\Phi(S_{T}) = (S_{T} - K)^{+}$$

</li>

<li>

Definition: $$\Gamma = \frac{\partial^2 V_{0}}{\partial S_{0}^2}$$

</li>

<li>

Analytic expression: $$\Gamma = \frac{e^{-\frac{d^2}{2}}}{S_{0}\sigma\sqrt{2\pi T}}$$
where: $$d = \frac{\ln\left(\frac{S_{0}}{K}\right) + (r + \frac{\sigma^2}{2})T}{\sigma\sqrt{T}}$$

</li>

<li>

Malliavin expression: $$\Gamma = \frac{e^{-rT}}{\sigma T S_{0}^2}\mathbb{E}\left[\Phi(S_{T})\left(\frac{W_{T}^2-T}{\sigma T} - W_{T}\right)\right] $$

</li>

</ul>

<h3><b>b) Python Implementation</b></h3>

```python
def MonteCarloEuropeanCallGamma(r,sigma,S0,K,T,N):
    start = time.time()
    d = (np.log((S0) / (K)) + (r + (sigma**2 / 2)) * T) / (sigma * np.sqrt(T))
    SAMPLE = []
    PATH = []  
    for i in range(N):
        W = np.random.normal(0, np.sqrt(T))                                             
        S = S0*np.exp((r - (sigma**2) / 2)*T + sigma*W)                                 
        V = max(S - K,0)
        SAMPLE.append(((np.exp(-r*T)) / (sigma * T * S0**2)) * V * (((W**2 - T) / (sigma * T)) - T))
        PATH.append(sum(SAMPLE) / len(SAMPLE))
    end = time.time()   
    print("Risk-free interest rate: r = " + str(r))
    print("Volatility: \u03C3 = " + str(sigma))
    print("Initial asset price: " + "S0".translate(str.maketrans("0", "₀")) + " = " + str(S0))
    print("Strike price: K = " + str(K))
    print("Maturity date: T = " + str(T))
    print("Number of simulations: N = " + str(N))
    print("Computation time: " + str(round(end - start, 3)) + " seconds")
    plt.rcParams["font.family"] = "Times New Roman"
    pd.Series(PATH).plot(linewidth = 0.75, color = "b")    
    plt.axhline((np.exp(-(d**2) / 2)) / (S0*sigma*np.sqrt(2*np.pi*T)), color = "black", linestyle = "--", linewidth = 0.75)                
    plt.xlabel("Number of Simulations")
    plt.ylabel("Estimated Gamma")
    plt.legend(["Sample Mean", "Exact Value (" + str(round((np.exp(-(d**2) / 2)) / (S0*sigma*np.sqrt(2*np.pi*T)),4)) + ")"])
    plt.title("Monte-Carlo Simulation of the Gamma of a European Call Option")
    plt.rcParams['figure.dpi'] = 300
    plt.rcParams['savefig.dpi'] = 300

MonteCarloEuropeanCallGamma(0.1,0.2,100,100,1,20000)
```

<h2><b>4) Monte-Carlo Simulation of the Delta of an Asian Fixed Strike Call Option</b></h2>

<h3><b>a) Theoretical Results</b></h3>
<ul>


<li>

Payoff function: $$\Phi(S_{T}) = \left(\frac{1}{T}\int_{0}^{T}S_{t}\mathrm{d}t - K\right)^{+}$$

</li>

<li>

Definition: $$\Delta = \frac{\partial V_{0}}{\partial S_{0}}$$

</li>

<li>

Malliavin expression: $$\Delta = \frac{e^{-rT}}{S_{0}\sigma^2}\mathbb{E}\left[\Phi(S_{T})\left(\frac{2(S_{T} - S_{0})}{\overline{S}_{T}} + \sigma^2 - 2r\right)\right]$$
where:
$$\overline{S}_{T} = \int_{0}^{T}S_{t}\mathrm{d}t$$

</li>

</ul>

<h3><b>b) Python Implementation</b></h3>

```python
def MonteCarloAsianFixedStrikeCallDelta(r,sigma,S0,K,T,M,N):
    start = time.time()
    dt = T / M
    SAMPLE = []
    PATH = []
    S = np.zeros(M + 1)
    S[0] = S0
    for i in range(N):
        for j in range(1,M+1):
            S[j] = S[j-1]*np.exp((r - (sigma**2)/2)*dt + sigma*np.sqrt(dt)*np.random.normal(0,1))
        STBar = np.mean(S[1:])
        VT = max(STBar - K, 0)
        SAMPLE.append(((np.exp(-r*T)) / (S0*sigma**2)) * (VT*(((2*(S[-1]-S0)))/(STBar)+sigma**2-2*r)))
        PATH.append(sum(SAMPLE) / len(SAMPLE))
    end = time.time()
    print("Risk-free interest rate: r = " + str(r))
    print("Volatility: \u03C3 = " + str(sigma))
    print("Initial asset price: " + "S0".translate(str.maketrans("0", "₀")) + " = " + str(S0))
    print("Strike price: K = " + str(K))
    print("Maturity date: T = " + str(T))
    print("Number of monitoring times: M = " + str(M))
    print("Time increment between monitoring times: \u0394t = " + str(round(dt,3)))
    print("Number of simulations: N = " + str(N))
    print("Computation time: " + str(round(end - start, 3)) + " seconds")
    plt.rcParams["font.family"] = "Times New Roman"
    pd.Series(PATH).plot(linewidth = 0.75, color = "b")                         
    plt.xlabel("Number of Simulations")
    plt.ylabel("Estimated Delta")
    plt.legend(["Sample Mean"])
    plt.title("Monte-Carlo Simulation of the Delta of an Asian Fixed Strike Call Option")
    plt.rcParams['figure.dpi'] = 300
    plt.rcParams['savefig.dpi'] = 300

MonteCarloAsianFixedStrikeCallDelta(0.1,0.2,100,100,1,252,20000)
```

<h2><b>5) Monte-Carlo Simulation of the Gamma of an Asian Fixed Strike Call Option</b></h2>

Payoff function: $$\Phi(S_{T}) = \left(\frac{1}{T}\int_{0}^{T}S_{t}\mathrm{d}t - K\right)^{+}$$

<li>

Definition: $$\Gamma = \frac{\partial^2 V_{0}}{\partial S_{0}^2}$$

</li>

where:
$$\overline{S}_{T} = \int_{0}^{T}S_{t}\mathrm{d}t$$

</li>

</ul>
